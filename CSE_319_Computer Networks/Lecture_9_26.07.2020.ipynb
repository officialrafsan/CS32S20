### TCP

![image.png](figures/fig181.PNG)

transport layer is reliable channel but when we use internet network layer is used which is unreliable. TCP provides reliability on unreliable network channel. From these figs below we can see how. transport layer is process to process communication and network layer is end to end communication. Network layer isn't reliable due to hardware medium. In the following lectures we'll see how we can use some protocol to make network layer reliable. These protocols are RDT 1.0, 2.0, 2.1, 2.2, 3.0 .

![image.png](figures/fig182.PNG)

![image.png](figures/fig183.PNG)

When we are to make the unreliable channel reliable the complexity of rdt will depend on the characteristics of the unreliable channel (network channel).

![image.png](figures/fig184.PNG)

- rdt_send() : this function is called by application layer, that's why it is called from above. This function works in transport layer. This function's job is to initiate rdt protocol.
- udt_send() : this function is called by  rdt to transfer the packet via unreliable network channel.
- rdt_rcv() : this function is called when a packet arrives at receiver side. This maintains rdt protocol in receiver side. i.e. send acknowledgement to receiver etc.
- deliever_data() : called  by rdt to deliver data to application layer when the data packet is healthy and error/bug free.
