In rdt 1.0 we assumed that the channel is reliable. But in reality the channel may not be reliable. So to deal with this reliability issue we'll use rdt 2.0, 2.1, 2.2, 3.0. Each iteration is improved from the previous. Let's start from **rdt 2.0**

![image.png](figures/fig188.PNG)

human analogy : when we can hear the other person we reply that we heard it but when we can't then we say didn't hear it then other person says again. This is how human recover from data loss.

We can use the same method in data transfer operation to recover from bit error or other types of error cause by unreliable network channel. If the packet is received without error then the receiver sends positive acknowledgement(ACK) otherwise sends negative acknowledgement(NAK). If the sender gets ACK then it sends the next packet otherwise sends the same packet again. This protocol is rdt2.0 .

![image.png](figures/fig189.PNG)

Now let's see the FSM of rdt2.0 and how does it work. So in in sender side FSM we can see that first we wait for call from above. when rdt_send() is called then we send the packet to receiver end and wait for acknowledgement, upon receiving acknowledgement we either send the packet again if we get NAK and again wait for acknowledgement or wait for call from above when we get ACK. And in the receiver side we wait for call from below we return acknowledgement and then again wait from call from below.

Now the major drawback of rdt2.0 is what if the acknowledgement is corrupted. Like the receiver sent ACK but sender received NAK. Now the sender will resend the packet, this will result in duplicate data in receiver side. This problem will be solved in rdt2.1

![image.png](figures/fig193.PNG)

**rdt2.1**

We can recover from the flaw of rdt2.0 by adding a sequence number(0,1) to the data packet. This will work like sender will send a packet with sequence number 0 (suppose) upon receiving, the receiver sends ACK but its corrupted so the sender again sends the same packet with sequence no. 0. Now the receiver will see there are two healthy packet(if the 2nd packet arrives without error) with sequence no. 0 so it'll keep one of the packet and discard another one and again send ACK. Upon receiving ACK the sender will send the next packet with sequence no. 1. This is the protocol of rdt2.1 by using this we can overcome rdt2.0's fetal flaw.


Let's see FSM of rdt2.1

Sender side 

Here in qst state sender wait for call 0 from above (0 sequence no.) when called sender sends packet with 0 sequence then wait for ACK or NAK of 0 sequence packet in 2nd state. If NCK then again send packet otherwise move to 3rd state. In 3rd state packet with sequence 1 is send and again goto 4th state and wait for ACK or NAK of 1 sequence packet If NCK then again send packet otherwise move to 1st state

![image.png](figures/fig194.PNG)

Receiver side 

Receiver wait for call from below for 0 sequence receiver receives it check for error if corrupted+ expected sequence not matched then sends NAK and wait in 1st state. If packet not corrupt then send ACK and goes to 2nd state. In 2nd state same thing happens. 

![image.png](figures/fig195.PNG)

The problem of rdt2.1 is receiver doesn't know that the acknowledgement it sent to sender was received without error at sender side and another thing is rdt2.1 is complex due to there are two acknowledgement ACK & NAK this results in too much state. To resolve this issue we'll use rdt2.2

![image.png](figures/fig196.PNG)


In **rdt2.2** NAK is removed. The protocol works like receiver sends ACK with the sequence number of last packet. Like sender will send a packet with sequence number 0 and receiver will add the sequence number 0 with ACK. If the packet is corrupted then receiver won't send any ACK. Same method is used for next packet with sequence number 1. This is how rdt2.2 works. But there's a problem with it. The problem is as in this protocol there's no NAK so how would sender know that if the packet was sent without error or as receiver won't send any NAK so how long sender will wait for that acknowledgement. This problem is solved in rdt3.0

In **rdt3.0** sender will wait for fixed amount of time. If in this time sender doesn't get any acknowledgement then sender will retransmit the packet. Sender will send each packet with sequence number so duplicate data can be handled in receiver side as we have seen in rdt2.1 . It will need a timer to count RTT. Here we can see that 1Gbps line utilization is very poor. So to utilize the line more we can use pipeline. It'll work like previously sender sent one packet at a time and only used sequence 0,1..But now sender will send 3/4/..n packets at a time or more depending on the line bandwidth with sequence 1,2,3,..,n depending on the number of packets. And the acknowledgement will be of also n packets. As we are sending set of packets so there will be buffering, or packet/ACK out of order.
