#### Today's agenda is learning about Web and HTTP.

![image.png](figures/fig76.PNG)

In our day to day life we use web alot. And web runs on http protocol. HTTP protocol use TCP cause in web browsing we can't consider loss. Reference objects are known by URL( Uniform Resource Locater). 

![image.png](figures/fig77.PNG)



Let's see how HTTP works. Basically HTTP follows client server architecture. That means on server side theres a software running on hardware i.e. apache server(open source software) or IIS(window's server software) most of the web service use apache and on the client side we use software like chrome/opera to browse the internet. In client server architecture client initiates the communication. So in this case we'll sent a request from browser to access a site like www.google.com. This is a http request we sent this request to server and server returns a http response that's why we can see google's homepage. This is HTTP protocol.

![image.png](figures/fig78.PNG)



client initiate tcp connection to server **at port number 80.** Ip address and port number, together are considered as socket, and **tcp port number is 80**. On the server side if we want to send a request we send to the http socket which is server side ip address and port number will be 80. HTTP is stateless that means when you send a request to sever, server doesn't store the information/record about who sent the request(normally but we'll see it also keeps the states). Protocols that maintains states are complex cause it has to maintain past history resulting in slower browsing speed.

![image.png](figures/fig79.PNG)



**HTTP connections:** HTTP has two types of connections one is called persistent and another one is non-persistent. When we sent a http request to access a web site we may have to establish multiple tcp conneection in non-persistent. Cause a web page can have many objects, to retrieve those objects in a webpage we need tcp connection but in non-persistent connection system after retrieving each object tcp connection is closed and another tcp connection has to established to retrieve another object. So if a webpage has N objects then tcp connection needs to established N times. On the other hand regardless of the number of objects we can retrieve all the objects of the webpage with a single tcp connection making it more faster. This is the main difference between these two.

![image.png](figures/fig80.PNG)
![image.png](figures/fig81.PNG)
![image.png](figures/fig82.PNG)



RTT(Round Trip Time) is the time it takes to sent a request to server + the time to get the response from the server. In non persistent system to get a single object 2 RTT is needed. 1st RTT is for initiating connection and 2nd one is to get the object. So in non-persistent HTTP response time of a single object is 2RTT + file transmission time. And if there's N objects in the webpage then total response time to get the whole webpage is (N+1)x(2RTT + file transmission time). Here we say N+1 cause first we need to setup connection to retrieve the html page then we retrieve N objects of the page.

![image.png](figures/fig83.PNG)



In persistent after connecting using TCP only takes 2RTT+file transmission time(of html page) after that we only consider 1RTT + file transmission time(of objects) cause server keeps connection open until the the transfer is over then closes the connection.

Previously we got to know that web server doesn't keeps any state. But here we can see sever keeps state known as cookies. So whats cookie? When we browse we sometimes see webpage like amazon prompts for permission to keep your browsing history/cookie cause they want to know what you like to see and can suggest similar products. They keep records of your browsing history against your ip address/unique primary key/login(if you login to those sites they can keep track of you browsing history easily) etc. To maintain this cookie both client and server maintains some tasks/components. There are four components.<br>**Explanation** When we visit amazon for the first time cookie header line will be blank cause we are new viewer, but when we visit next time there would be a cookie header in next http request cause there's a unique id assigned to each of us against our ip address/unique primary key/login_info etc. Cookie files are kept on host end and in back end database. Cookie files contains our browsing records like for amazon which product we visited/liked, which product we searched more, which product we order. They keep this cookie files to improve their business by recommending similar products.

![image.png](figures/fig92.PNG)



For this example lets consider our browser has a cookie file containing id of ebay now when we visit amazon there's no id for it so amazon creates a id and stores it in it's database then return the id with the response and our browsers update the cookie file with our amazon id. So next time when we visit amazon there's a cookie id for us so amazon can identify us, keep track of our visiting records and suggest products that we like.

Now if we clean cookie from our browser does it benefits us? Actually it doesn't. Normal concept can be like when we clean the cookie amazon shouldn't be able to recognize us and would give us new id and our browser will update the cookie file. But the reality is amazon can still identify us by our ip address or browsers address or if we login to it, then it sends our original id and browser saves it. Thats why clearing cookie doesn't benefit us.

![image.png](figures/fig93.PNG)


Drawbacks of cookie is they can learn a lot about you resulting in privacy breach . Thats why websites asks for permission. So if you give permission then you can't claim that they breached your privacy.

Web cache (proxy server) is a copy of original server. Like when we give proxy attendance for our friends. They are not physically in the class but we give their attendance. Proxy server copies the content of original server and gives proxy of the original server. Why there's a need for proxy server? Assume you want to watch a video the youtube server is across the sea so it'll take time to retrieve the video and the connection will be slow. So if we copy the content of the youtube server and make a proxy server in Bangladesh then you can watch it with no buffer. Thats the reason watching videos in youtube is faster than watching videos in other platforms. Now you may think that youtube has a huge data center/server how can a single server in Bangladesh can copy the content of that. Actually proxy servers tries to copy as much data as it can in multiple servers that can combinedly act as a proxy server. We also use proxy server to reduce the load on access link(we'll see this in the example of caching)

![image.png](figures/fig95.PNG)
![image.png](figures/fig96.PNG)



Lets see a caching example. Assume that you want to access a webpage containing some objects and each object size on an average is 100bits, and in 1 sec we can send 15 requests(avg rate from browaer to origin server 15/sec), so data rate is 100x15 = 1500Kbps =1.5Mbps. RTT between router and origin server is 2sec and access link speed/rate is 1.54Mbps. When there's more traffic in the access link performance falls. So each time we want to reach origin server we increase traffic on the access link, our lan connection has 1Gbps speed but its  utilization is only at (1.5/1024)x100 = 0.15% utilization cause we are using access link to goto origin server. Thus access link utilization is (1.50/1.54)x100 = 99%  reulting in great delay. We can calculate this delay. Total delay = internet delay(RTT)+access delay+lan delay = 2sec+minutes(due to utilization this delay will be in minutes) + micro seconds(due to less utilization this delay will be in micro seconds.) So we can assume that the delay is more than a minute. **How can we improve this delay?** 

![image.png](figures/fig97.PNG)

One solution is, we can increase the access link speed by 100times making it 154Mbps. So now the utilization is at 9.9%.

![image.png](figures/fig98.PNG)

Another solution can be make a local web cache. Now we need to calculate the link utilization and total delay. If we see that still the utilization is high then we need to increase link bandwidth/speed and if the utilization is considerable we can keep continue with the cache. Web cache is cheap comparatively cause it only takes couple lakhs tk to install a cache but increasing link speed is expensive monthly expenditure can be couple of lakhs tk. So making a local cache server is more financially beneficial.

![image.png](figures/fig99.PNG)

Lets do some calculation. Suppose the request we send can be satisfied 40% at cache and 60% data needs to be retrieve from origin server. Data rate to browsers over link is 0.6x1.50Mbps = 0.9Mbps. So link utilization is 0.9/1.54 = 0.58 = 58%. So total delay is about 1.2sec. So here we can see that the response time is less then RTT. So just installing a cache server can significantly improve time. So no need to increase link speed. But if you want to reduce time further than you can increase link speed.
