# Protocol
![image.png](figures/fig5.PNG) ![image.png](figures/fig6.PNG)<br>A network protocol is an established set of rules that determine how data is transmitted between different devices that are connected to network. Essentially, it allows connected devices to communicate with each other, regardless of any differences in their internal processes, structure or design.

# Host : Calculating speed of data packets
![image.png](figures/fig15.PNG)<br>
Here L is the size of each packet and R is the bandwidth/transmission rate. So packet transmission delay is L/R. Packet transmission delay is the total time needed for transmitting a packet considering propagation delay is 0. So lets say we have N packets then the delay would be Nx(L/R). Let L=1MB and R=2MB so to transfer 100MB data how much time it would take? ans is 100 x 1/2 = 50sec.<br>

# Network Core
![image.png](figures/fig19.PNG)<br>
Network core or internet is like a mesh connection between routers. mesh is like all the components are connected to each other. In packet switching hosts beark the data into some packets and forward the packets across links on path from source to destination. Each packet is transmitted at full link capacity.<br>
![image.png](figures/fig20.PNG)<br>
In this fig we can see that there is one intermediate device between source and destination that is a router with propagation delay of 0. So there are two edges in this path. So to get the total delay <br> First we calculate 1 hop delay that is L/R = 7.5/1.5=5sec. So total duration of packet 1 to reach destination is 5 x 2(as there are two edges) =  10 sec.<br>
#### Total delay = 1 hop delay x no. of packet + no. of intermediate device x 1 hop delay = 5 x 3 + 1 x 5=20sec.
<br>**Explanation :** Cause when packet1 reaches the router packet2 is ready to be transmitted, when packet1 reaches destination packet2 reaches router and packet3 is ready to be transmitted when packet2 reaches destination packet3 reaches router and it takes another hop for packet3 to reach destination.<br>

**Example :**

Consider that the University of Asia Pacific (UAP) creates 843.75 GB data every 24 hours and needs to transfer to the server which is not located on the campus. In addition, the packet size for data transmission is fixed into 2MB, and the transmission rate between UAP and the server is 4MB per second.
How much time will it take to send all the data to the server? Consider that the communication has not any propagation delay.

Ans: 

Size = (843.75 x 1024) = 864000<br>
Total packets = 864000/2 = 432000<br>
Total delay = 432000 x (2/4) =  216000 seconds<br>

It will take total 216,000 seconds to transfer all the data( 843.75 GB) from uap to server<br>
