# Distributed Hash Table(DHT) :

To search for torrent we need to manage indexing as trackers aren't fixed due to churn. We can maintain indexing using hash table. Hash is function that return a particular output for a particular input. Output for that input is always same/fixed, `f(x)=y`, this ensures data integrity. Hash is one way, you can only get y by hashing x not the other way around. DHT will maintain indexing.

![image.png](figures/fig137.PNG)
![image.png](figures/fig138.PNG)

It's easy to search on numeric values than character value and as the key value is unique so there's no data redundancy. When we search a movie it's name is first hashed to get the unique key and using that key the back-end searches for the ip address associated with that key.

![image.png](figures/fig139.PNG)

![image.png](figures/fig140.PNG)

Let there are 8 peers. From a key we'll see the next peer. So if the key is 51, 60 is the closest immediate successor id so we assign this id to that key as the value of that key. Now if key is 60 than peer id is also 60. But if the key is 61 than the peer id would be 1. Cause the peer list is circular so after pee_id 60 comes peer_id 1 means next peer id from 60 is 1. That's why we assign peer id 1 to the key 61.

![image.png](figures/fig141.PNG)

Each peer is aware of it's immediate predecessor and successor.

![image.png](figures/fig142.PNG)

`Resolving query`. What is the value associated with key 53? By seeing the fig we can tell that the id would be 60. But let's see how does the system calculates it. Let peer 12 asked the query so immediate successor to it is peer with id 13 so peer 12 will ask peer id 13, but 53>13 so peer id 13 will pass this query to it's successor that is peer id 25. Still 25<53 so peer 25 will pass it to 32, 32 will pass it to 48, 48 pass it to 60. Now for peer id 60>53 so the so this peer's id (60) will be returned as the answer of the query. To complete this whole operation we need 0(N)  messages on average to resolve the query when there are N peers. 

![image.png](figures/fig143.PNG)

But if there are many peers lets say 1 million then the query would take 0(1 million) messages. So it's a huge complexity. We can reduce this complexity by using shortcuts. In this fig below we can see that the query took half the messages compared to before, also the complexity is reduced to 0(log N).

![image.png](figures/fig144.PNG)

**Peer Churn**

A particular peer know it's predecessor and successor. So it pings them after a certain period of time to know if they are alive or not. If they aren't alive then the peer selects the next (from the dead peer) peer and selects it as it's immediate successor. Like when 5 leaves 4 sets 8 as it's immediate successor. And when the dead peer comes alive then it sends a message to all other peer to tell that it is alive. Then peer 4 will again update it's immediate successor from 8 to 5.
