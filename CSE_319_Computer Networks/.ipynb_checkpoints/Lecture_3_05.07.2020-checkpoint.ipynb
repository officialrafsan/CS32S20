To communicate we need a app its called network app like meet, email etc. Multi user network games are like pubg , cod mobile etc.

Network app must be platform independent like it should be able to run in any platform. These apps are only made for end devices.

There are two types of architecture. Client server example is like FTP server, google searching etc and P2P example is like google meet, torrent where both parties can contribute/provide service.


![image.png](figures/fig64.PNG)

Server is always on to serve a client at any moment. When necessary only then client connects to get service. In client server architecture server serves the request of a client.

![image.png](figures/fig65.PNG)

On the other hand P2P architecture is where both of the clients can serve each other. Like in google meet both parties can present or share their knowledge. In torrent one can download and can do upload. So it's basically an architecture where client can also act like a server. 

![image.png](figures/fig66.PNG)

no loss means any kind of loss will not be tolerated. Elastic mean the throughput can vary with time. UDP is maybe used in platform where loss is elastic or considerable.

![image.png](figures/fig71.PNG)

TCP adds much overhead than UDP so in low speed streaming UDP is sometimes used.Different communications demand different protocols. We know about TCP and UDP protocol. TCP transfers data with extra care and it is highly reliable. On the other hand, UDP isn't reliable as it has fewer headers compared to TCP but it's faster cause it has less header. Now coming to the question. Assume we want to send an email. Can we tolerate packet loss? We can't. So we need a reliable protocol to transfer data. That's why TCP will be used in this communication. Again assume we need to watch a youtube video. In this scenario we focus more on transmission speed then reliability cause losing a few packets doesn't bother much, we can tolerate such negligible packet loss. In this scenario, as we focus more on transmission speed so we can use UDP. TCP can also be used but UDP is much faster.
These two scenario demands different protocols.
