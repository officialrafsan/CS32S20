**Attacking DNS**

- DDoS(Distributed Denial of Service):
    In this bot attacks the server with huge amount of queries, for that reason when the server gets new query it denies the query as it is very busy to process the queries made by the attacker. 
    - Bombard root server : Till date attacking root isn't successful, cause the attack would start from local DNS and local DNS in general doesn't goto root address as it has cached TLD server addresses. Local DNS bypesses root DNS. DNS also maintains traffic filtering. That is generally this attacks comes through ICMP, so if we close the ICMP port then the server will deny the attackers query making it more secure. This is traffic filtering. So root bombard isn't possible.
    
    - Bombard TLS server : TLD server is cached in local DNS server, so we can flood the TLD DNS server or authoritative server with huge amount of query to take down the TLD/authoritative server. So bombarding TLD is more dangerous.
    
    - Redirect attacks : In this we can intercept a query when it returns from TLD/authoritative and replace the ip address with a fake one, so when the user gets the ip he is redirected to a fake webpage in which if he gives his credentials his information can be hacked. Another thing we can do is intercept the packet and replace it with a bogus packet before it reaches TLD , so that the user can't get access cause TLD will return nothing for the bogus request This is DNS poisoning.
    
    - Exploit :  Here we exploit the the DNS like when a new query is made local will get it from TLD and will cache it so if we intercept the packet while it returns to local from TLD and replace the real ip with ours than local DNS will cache our ip for that particular domain name as it will think it's a reply from TLD. This is also a security breach.

# P2P architecture

In client-server architecture a server is always on and client can will be on when it requires something. here client initiates the connection. There's another type of architecture called peer to peer(P2P) architecture. The connection between peers is called P2P architecture. So what's a peer? A host that acts like both client and server is called peer. A peer can be a client or it can be a server to serve others depending on the need. A peer may not be active/on all the time this also refers that when a peer acts like a server it may not be on all the time. Ex of P2P is torrent(app soft). Torrent has both client and server side to it. As peers comes and leaves the network so they are intermittently connected and change ip address every time.

![image.png](figures/fig128.PNG)

Let's see an example.

Here we can see a p2p network with N peers and abundant/sufficient bandwidth. There's also a server containing a file of size F. Now the question is how much time it will take to distribute file from one server to N peers?

Server upload speed to network is $u_s$ , $i_{th}$ peer's download speed is $d_i$ and upload speed is $u_i$.

![image.png](figures/fig129.PNG)

`client - server approach`

server transmission must sequentially send/upload N file copies, so time to sent one copy is $\frac {F}{u_s}$ so to send N copies is $\frac {NF}{u_s}$ client download rate is $d_{min}$ so minimum download time is $\frac {F}{d_{min}}$ . So maximum time delay to distribute file of size F to N clients using client-server approach is $D_{c-s}\geq max(\frac {NF}{u_s}, \frac {F}{d_{min}})$. So the maximum delay can be the maximum upload time and maximum download time. This delay increases linearly as N increases. Cause download time is same for all clients are same but upload time for a server increases as it has to upload to more clients.

![image.png](figures/fig130.PNG)

`P2P approach`

In P2P a peer can work as a client and server. In this architecture the server must upload at least one copy of the file then 1st peer can download it, and when a 2nd peer wants to download the file it can download it from both server and 1st peer(half from server half from peer 1 and when a peer finishes download it acts like a server so it can also upload the file), then when a 3rd peer comes it can download the file from server, peer 1, peer 2. This goes for peer N. peer N can download from server and N-1 peers(if N=9 then 9th peer can download from server and previous 8 peers). 

Now time to upload one copy is $\frac {F}{u_s}$ , minimum download time is $\frac {F}{d_{min}}$ , maximum upload(to new peer) rate is ${u_s}+\sum {u_i}$ . So to distribute a file size of F to N clients using P2P architecture is $D_{P2P}\geq max(\frac {F}{u_s}, \frac{F}{d_{min}}, \frac{NF}{{u_s}+\sum {u_i}})$


![image.png](figures/fig131.PNG)


As a new client can download from multiple peers so the upload rate(upload file for the new client) of those peers are distributed among them. So increasing the number of peer will increase upload rate and decrease upload time. After the new client finishes download it also acts as a server.
