**FTP(File Transfer Protocol):**

FTP is based on client server architecture. Today file server is served via means of http protocol. But in classic file server system it runs on port 21 and follow ftp. In ftp files are kept in server and clients can access those files. Client initiate connection and server will serve the request. 

![image.png](figures/fig102.PNG)


Two sides of ftp are control and connection. Both sides have different port numbers. when server receives file transfer command, server opens 2nd TCP data connection (for file) to client this means when the server opens 2nd tcp data connection you can upload to server or download from server via port 20 then it closes. If you want to transfer another file then again server opens another tcp connection. FTP server maintains state for authentication. Like we don't need authentication to browse web but to browse a person's/company's movie server authentication is needed.

**DNS**

To access something we need to identify it first. But identifying via ip isn't possible for us we can only remember name. DNS does this conversion for us. We give name of webpage we want to access then this request goes to DNS, DNS converts it to ip address and sends it to our pc then our pc can access that webpage. DNS is actually a distributed database implemented in hierarchy setup. The main task of DNS is to convert name to ip. Its complexity depends on network edge.

![image.png](figures/fig114.PNG)

We don't centralize DNS for the reasons below:
* Single point failure: If centralized then when the DNS server fails we won't be able to access internet as we don't know the ip addresses of sites we want to visit.
* Traffic volume: If the DNS is centralized then there would be aa huge traffic load cause request will come from all over the world. Thats why it's not centralized.
* Distant centralized database : If centralized then it will take more time to get the ip converted from name
* Maintenance : It would be hard and costly to maintain a centralized DNS server. \
* Scale : If centralized then you can't make it bigger or smaller this means that centralized DNS doesn't scale.

These are the reasons for not using centralized DNS server.

DNS name resolution can be done using 2 ways. One is iterated query and another is recursive query. Lets discuss iterated query. Assume you want to access cis.poly.edu this web site. Now this request goes to your local DNS, but local DNS doesn't have this in its database so it sends the request to root dns, root dns replies to check in TLD dns, so local dns reaches tld and tld replies to check authoritative dns then local goes to authoritative dns and authoritative response with the site's ip address.

![image.png](figures/fig120.PNG)

In recursive local request root, root then goes to tld and tld goes to authoritative. Authoritative returns webpage's ip then tld returns this to root and root return this to local. Finally local returns the ip to you.

![image.png](figures/fig121.PNG)

DNS caching is more like proxy server. It stores ip from another server for a period of time so that it can give proxy. In proxy server we saw data was copied from origin server in DNS caching ip address with domain name is copied from origin server for a period of time.
