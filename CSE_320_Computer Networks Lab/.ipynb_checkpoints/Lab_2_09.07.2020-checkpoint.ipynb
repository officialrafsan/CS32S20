Lets design the logical side for this 2 networks. Switches belong to layer 2 that is data link layer and routers belong to layer 3 that is network layer. Switch forwards data in LAN, router forwards data in WAN.

![image.png](figures/fig8.PNG)


This is a client server network architecture. So we need a server and client are the pcs.

![image.png](figures/fig9.PNG)

To provide web service we need a DNS server with http server.

![image.png](figures/fig10.PNG)

We need to specify the the devices with unique ip address. We can do this by using static or by dhcp. Static is hard coded and Dynamic Host Configuration Protocol (DHCP) is dynamic. Let there's 250 pc in a network hardcoding all of them is very difficult so we can use dhcp where it will auto configure the network.

Here we add DNS and DHCP servers. DNS servers helps to decode webpage name and return ip address and DHCP server helps us to configure all the end devices connected to the network.

![image.png](figures/fig11.PNG)


![image.png](figures/fig12.PNG)


* DHAKA ZONE (192.168.1.0):
          Broadcast : 192.168.1.255
          Usable : 192.168.1.1 ~ 192.168.1.254
          Default gateway : 192.168.1.1 (Router5 Fa0/0) {generally we set the first address as default gateway)
          Ip for Server : 192.168.1.245~254 {reserved for server}
          Ip for client : 192.168.1.2~244
          {We need to statically specify all the servers}
          DHCP_DHK : 192.168.1.254
          DNS_DHK : 192.168.1.253
          Google : 192.168.1.252
          Yahoo : 192.168.1.251
          UAP : 192.168.1.250
          
          
          

* CTG ZONE (192.168.2.0):
          Broadcast : 192.168.2.255
          Usable : 192.168.2.1 ~ 192.168.2.254
          Default gateway : 192.168.2.1 (Router5 Fa1/0)
          Ip for Server : 192.168.2.245~254 {reserved for server}
          Ip for client : 192.168.2.2~244
          DHCP_CTG : 192.168.2.254
          DNS_CTG : 192.168.2.253
          Amazon : 192.168.2.252
          Daraz : 192.168.2.251
          Evaly : 192.168.2.250
          




## 1. We need to specify DHCP servers ip statically.

**Here I select an ip address from the addresses that were reserved for servers. And also give DNS ip address.**

![image.png](figures/fig13.PNG)


## 2. Then we need to configure DHCP server. 

**First goto services then turn it on. After that give the default gateway ip and DNS server's ip than give the start address. In our case the start address for this network is 192.168.1.2 cause we already selected 192.168.1.1 as default gateway and 192.168.1.0 is reserved.**

![image.png](figures/fig14.PNG)

## 3. End device ip config

**We goto each end device (here its pc) and goto desktop->ip config-> then just select dhcp shown below and it will auto config itself.**

![image.png](figures/fig15.PNG)


## 4. Configure each server

**Each server needs to be configure individually and you have to give each server its ip address statically as show below. Then if you want to change the webpage it shows goto services->http-> then turn it on(by default its turned on) and edit the index.html file.**

![image.png](figures/fig16.PNG)
![image.png](figures/fig16_a.PNG)


## 5. Configure DNS. 

**Now you need to configure your DNS server. To do that give it , its ip address and the default gateway then goto services->DNS-> turn it on and add the domain name and ip address of that server. For for the uap server I gave it its domain name www.uap.com and the uap server's ip address 192.168.1.250. Then click add or save. Shown as below.**

Function of DNS_Server : To browse the web we give web address like  www.uap.com, we don't give ip address. So to access this web address first this request is sent to DNs server then the DNS server returns the ip address associated with this web address to the end device. Than the end device send request via this ip address to gain access to the webpage. 

![image.png](figures/fig17.PNG)
![image.png](figures/fig18.PNG)


## 6. Check if connection is ok then move on to CTG network. and repeat the same process there.

**To check if the connection is ok or not , select a pc then goto Desktop->Web-> and type in the domain name that is entered in DNS server and has a ip address. Doing this will return a page as shown below( i have edited the index.html file so mine looks a bit different). If it a page like this then the connection is ok.**

**N.B:** DNS server was added in both side to reduce delay. Like if there was a only one DNS in dhaka then end device from CTG will sent a request to dhaka DNS to get the ip of the website that it want to access. So its a lengthy process. To solve this we set a DNS in CTG so that we can get ip of desired website reducing the delay of getting the ip address.

![image.png](figures/fig19.PNG)
