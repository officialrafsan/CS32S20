**CIDR :** (review lab 4 for CIDR & VLSM)


2^0 = 1

2^1 = 2 - 2 = 0

2^2 = 4 - 2 = 2

2^3 = 8 - 2 = 6

2^4 = 16 - 2 = 14

2^5 = 32 - 2 = 30

2^6 = 64 - 2 = 62

2^7 = 128 - 2 = 126 (32-7)=25 

2^8 = 256 - 2 = 254

2^9 = 512 - 2 = 510

For 7 bit host network bit is 25 so binary representation is, 

255.255.255.128/25 : 11111111.11111111.11111111.1 0000000

Routing
------
Static: We did that in lab 3.

Dynamic: 

- RIP (Routing Information Protocol), Classfull IP addressing

- VLSM: RIP Version 2,Classless IP addressing

config steps:

1. We need to find the connected network addresses (show ip route)
2. configure in terminal (configure terminal)
3. router rip
4. version 2
5. no auto-summary
6. we will add all connected networks in the RIP

**We need to find the connected network addresses (show ip route)**

using `show ip route` command we can see all paths the router can send packets to. here c means connected.

![image.png](figures/fig77.PNG)

**configure in terminal (configure terminal)**
**router rip**
**version 2**
**no auto summary**
**we will add all connected networks in the RIP, using `show ip route` we got all the connected networks**


![image.png](figures/fig78.PNG)
