This lab is a continuation from last lab. So lets build 2netwrok in both sides connected through many interconnected routers. Our goal of this lab is to build the same layout as last lab but with multiple routers to get an idea of real life scenario.

![image.png](figures/fig22.PNG)


In the fig above we can see there are 10 networks. Now you must be thinking how come there are 10 networks. The answer is we count each connection line of the router as a network. Like router1 & router8 each has 3 connecting lines, so sum is 6 lines then router3 has a connecting line with router5 and router5 has a connecting line with router7 now the sum is 8 lines and finally router6 has connecting line with router4 and router4 has a connecting line with router2, now the sum is 10 connecting lines. So there are 10 networks in this architecture.


Now we need to make up a table so that we can configure the routers. So how do we setup this table? Lets start from router1. Router1 is connected to BD network via F0/0 and it's default gateway is 192.168.1.1 this ip belongs to network 192.168.1.0 . But in the table I write 192.168.1.0/24, why did I write it? because there are 3 network octates in class C, so 3x8=24. For the ease of writing I'll refer 192.168.n.0/24 just as 192.168.n.0 .

So router1 and router3 are connected in the same network via port Se3/0 and router1 will send packet to router3 via port Se3/0 as router3 is set to be the next hop. Router3 and router5 is connected in the same network via port Se2/0 and router3 will forward the packet to router5 via port Se2/0 as router5 is set to be the next hop. Similarly router5 will forward the packet to router7 and router7 will forward it to router8. Now this packet can goto India via port Fa0/0 and when India sends response the packet will come to router8, router8 will forward it ot router6, router6 will forward it to router4, router4 will forward it to router2 and router2 will forward it to router1 thus the response can goto BD via router1's port Fa0/0. This was the theory now lets configure the routers.
