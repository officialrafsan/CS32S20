## Routing Config (CLI + RIP V2)

config steps:
0. command : enable
1. We need to find the connected network addresses (show ip route), command : show ip route
2. configure in terminal, command : configure terminal
3. router rip, command : router rip
4. version 2, command : version 2
5. no auto-summary, command : no auto-summary
6. we will add all connected networks in the RIP, command : network n.n.n.n( here it is the ip address that we can see in show ip route and has a 'c' written before it.)
