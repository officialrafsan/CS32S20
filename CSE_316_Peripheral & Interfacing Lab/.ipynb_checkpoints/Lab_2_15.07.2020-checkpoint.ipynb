## PORT 1: 

Arduino is powered up by this pin. We can connect arduino with our system using this port. When we connect it with system it gets the needed voltage to work and we can upload our codes in arduino using this port. Without this port we can't upload our code.

![image.png](figures/fig2.PNG)

## PORT 2: 

Let we uploaded code in arduino and don't need to connect to system or we have to keep the arduino in a place where you can't connect it to your system then we can power up the arduino using this barrel jack.

![image.png](figures/fig3.PNG)

## Component 3: 

Voltage regular is used to stabilize power delivery. It helps to keep the voltage stable.

![image.png](figures/fig4.PNG)


## Component 4: 

Crystal Oscillator handles time related things. Like our code may have a wait/delay operation, so this operation is handled by the Crystal Oscillator.

![image.png](figures/fig5.PNG)

## PIN 5,17: 

These pins helps us to reset the operation of arduino so that it starts over the operations.

![image.png](figures/fig6.PNG)


## PIN 6,7,8,9 : 

We can output 3.3, 5v, GND is ground and using Vin you can also power up the board.

![image.png](figures/fig7.PNG)

## PIN 10 : 

A0-A5 are the analog pins. They are used when we work with analog signal.

![image.png](figures/fig8.PNG)

## Component 11 : 

Microcontroller is the processor of the board.

![image.png](figures/fig9.PNG)

## PIN 12 : 

We use this pin for serial programming.

![image.png](figures/fig10.PNG)

## Component 13 : 

By this led we can know that if the board is powered up or not.

![image.png](figures/fig11.PNG)

## Component 14 : 

By this led we can know that if the board is transmitting/receiving or not.

![image.png](figures/fig12.PNG)

## PIN 15 : 

In this section -3,-5,-6,-9,-10,-11 these pins are the pwm(pulse width modulation) pins. 

![image.png](figures/fig13.PNG)

Our task is to simulate a simple led blinking code. Steps are below 

1. Click on p and select arduino uno and then drag and drop it to the layout.

![image.png](figures/fig16.PNG)

![image.png](figures/fig17.PNG)

2. Again click on p and search for 'resistor pot' then select POT device and drag & drop it to the layout.

![image.png](figures/fig18.PNG)

![image.png](figures/fig19.PNG)

3. Changethe resistor value from 1k to 220 then click ok.

![image.png](figures/fig20.PNG)

4. Connect resistor to pic 13 of the board.

![image.png](figures/fig21.PNG)

5. Click on p search for " led" and select which ever led you like just see that it is in ACTIVE library. Drag and drop it to layout and connect it with the resistor.

![image.png](figures/fig22.PNG)

![image.png](figures/fig23.PNG)

6. Then we need to put a ground in our layout. We select ground like show in below then drag and drop it to our layout. Then we connect it to the led.

![image.png](figures/fig24.PNG)

![image.png](figures/fig25.PNG)

7. Now we need to open the code. Then verify it and copy the hex file path.

![image.png](figures/fig26.PNG)

8. Double click on the board and paste the copied hex file path in the program file field.

![image.png](figures/fig27.PNG)

7. Simulate it.
