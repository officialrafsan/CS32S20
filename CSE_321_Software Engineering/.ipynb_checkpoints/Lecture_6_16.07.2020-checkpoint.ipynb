## RAD [learn more](https://www.guru99.com/what-is-rad-rapid-software-development-model-advantages-disadvantages.html)

Though RAD and waterfall model has differences between them but in some aspects they are similar. RAD or Rapid Application Development process is an adoption of the waterfall model; it targets at developing software in a short span of time.

![image.png](figures/fig42.PNG)

This model focuses on element based construction. It has 5 phases

1. **Business Model :** A business purpose will be stated here.
2. **Data Modeling :** Collecting data from business modeling. Refining the data so that it becomes highly correlated with business.
3. **Process Modeling :** Modifying objects gained from business and data models.
4. **Application Generation :** Making the app, coding part.
5. **Testing & Turnover :** Testing before deployment, and deployment process.

![image.png](figures/fig43.PNG)

RAD is a sequential mode.It's like waterfall model in this aspect.

![image.png](figures/fig44.PNG)

![image.png](figures/fig45.PNG)

The biggest drawback of RAD is you need highly skilled designers.

![image.png](figures/fig46.PNG)

### Spiral Model [learn more](https://www.javatpoint.com/software-engineering-spiral-model)
[learn more](https://www.geeksforgeeks.org/software-engineering-spiral-model/)

![image.png](figures/fig47.PNG)

Spiral Model has the first risk management system. We can divide spiral model in 4 phases. It's very important for both exam and viva you can use any diagram from the 2 diagrams below.

![image.png](figures/fig48.PNG)

![image.png](figures/fig49.PNG)

**Sectors/Phases of spiral model**

1. **Objective setting :** Identifying objects, that is we discuss user need/description here.
2. **Risk assessment and reduction :** Find out what type of risks can occur in hardware or software side of the system and how we can reduce the risks.
3. **Development and validation :** Make the application, coding part.
4. **Planning :** Project is reviewed here and further planning is done here.



![image.png](figures/fig51.PNG)

Biggest advantage is we have risk management in this model. Biggest drawback is too much dependable risk analysis. Like the developers might leave leaving the project half done and new developer will take much time to understand.

![image.png](figures/fig52.PNG)

### Incremental Model [learn more](https://www.geeksforgeeks.org/software-engineering-incremental-process-model/)

It's also called successive model cause it's always being updated and newer version is more improved than previous one. Like we make the system with minimum requirement then customer give time to build the other parts of the model iteratively. Like version1 of a project may contain database and frontend, version2 may contain login system, version3 may contain server system so like this after N versions the system will be fully complete. So it's like an incremental model.

![image.png](figures/fig53.PNG)
![image.png](figures/fig54.PNG)
![image.png](figures/fig55.PNG)

Assume one team works on the whole project so after creating a subsystem then they can move on to the next one this is staged delivery model like after building database frontend then they'll build login system and so on. If you have multiple teams let say N team so you can build N subsystem at the same time this is parallel development model like one team can build database frontend another can build login system and N teams can build N subsystems.

### V Model [learn more](https://www.geeksforgeeks.org/software-engineering-sdlc-v-model/)

V model is the combination of two life cycle  models. One is SDLC and another one is STLC.

![image.png](figures/fig59.PNG)

V model is referred to as verification and validation model cause each step has it's own corresponding testing. The steps are sequential . We can see this in the fig below.

![image.png](figures/fig60.PNG)

1. **Module design** is designing a particular module like designing a login system. And testing this single module is called **unit testing.** 
2. **Architecture design** is designing the all the modules and after we design an architecture then we do integration testing, basically **integration testing** is testing all the modules together or we can say testing the whole architecture.

As this is a sequential model so after unit testing we can do integration testing. Above I just mention re relationship between design phase and testing phase that is which testing phase is for which design phase. But the working procedure is 1st the design phase will be completed then we'll do coding then testing phase will be executed. And the order of design phase is top to bottom and order of testing phase is bottom to top(marked with arrows).
