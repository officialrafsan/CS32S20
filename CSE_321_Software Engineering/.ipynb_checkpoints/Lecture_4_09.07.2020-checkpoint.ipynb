<h3>Phase 1: Requirement collection and analysis:</h3>
<p>The requirement is the first stage in the SDLC process. It is conducted by the senior team members with inputs from all the stakeholders and domain experts in the industry. Planning for the quality assurance requirements and recognization of the risks involved is also done at this stage.</p>
<p>This stage gives a clearer picture of the scope of the entire project and the anticipated issues, opportunities, and directives which triggered the project.</p>
<p>Requirements Gathering stage need teams to get detailed and precise requirements. This helps companies to finalize the necessary timeline to finish the work of that system.</p>
<h3><a></a>Phase 2: Feasibility study:</h3>
<p>Once the requirement analysis phase is completed the next step is to define and document software needs. This process conducted with the help of 'Software Requirement Specification' document also known as 'SRS' document. It includes everything which should be designed and developed during the project life cycle.</p>
<p><strong>There are mainly five types of feasibilities checks:</strong></p>
<ul>
<li><strong>Economic:&nbsp;</strong>Can we complete the project within the budget or not?</li>
<li><strong>Legal:</strong>&nbsp;Can we handle this project as cyber law and other regulatory framework/compliances.</li>
<li><strong>Operation feasibility:</strong>&nbsp;Can we create operations which is expected by the client?Like the client want a VR monitoring system. The technology exists and we can make it.</li>
<li><strong>Technical:</strong>&nbsp;Need to check whether the current computer system can support the software. Like the client want VR monitoring system but he is in a remote area where he can't use the recommended computer to run the software.</li>
<li><strong>Schedule:</strong>&nbsp;Decide that the project can be completed within the given schedule or not.</li>
</ul>
<h3><a></a>Phase 3: Design:</h3>
<p>In this third phase, the system and software design documents are prepared as per the requirement specification document. This helps define overall system architecture.</p>
<p>This design phase serves as input for the next phase of the model.</p>
<p>There are two kinds of design documents developed in this phase:</p>

1. High level design (HLD): It give the architecture of software product. 
2. Low level design (LLD): It describe how each and every feature in the product should work and every component. 


<p>High-Level Design (HLD)</p>
<ul>
<li>Brief description and name of each module</li>
<li>An outline about the functionality of every module</li>
<li>Interface relationship and dependencies between modules</li>
<li>Database tables identified along with their key elements</li>
<li>Complete architecture diagrams along with technology details</li>
</ul>
<p>Low-Level Design(LLD)</p>
<ul>
<li>Functional logic of the modules</li>
<li>Database tables, which include type and size</li>
<li>Complete detail of the interface</li>
<li>Addresses all types of dependency issues</li>
<li>Listing of error messages</li>
<li>Complete input and outputs for every module</li>
</ul>
<h3><a></a>Phase 4: Coding:</h3>
<p>Once the system design phase is over, the next phase is coding. In this phase, developers start build the entire system by writing code using the chosen programming language. In the coding phase, tasks are divided into units or modules and assigned to the various developers. It is the longest phase of the Software Development Life Cycle process.</p>
<p>In this phase, Developer needs to follow certain predefined coding guidelines. They also need to use programming tools like compiler, interpreters, debugger to generate and implement the code.</p>
<h3><a></a>Phase 5: Testing:</h3>
<p>Once the software is complete, and it is deployed in the testing environment. The testing team starts testing the functionality of the entire system. This is done to verify that the entire application works according to the customer requirement.</p>
<p>During this phase, QA and testing team may find some bugs/defects which they communicate to developers. The development team fixes the bug and send back to QA for a re-test. This process continues until the software is bug-free, stable, and working according to the business needs of that system.</p>
<h3><a></a>Phase 6: Installation/Deployment:</h3>
<p>Once the software testing phase is over and no bugs or errors left in the system then the final deployment process starts. Based on the feedback given by the project manager, the final software is released and checked for deployment issues if any.</p>
<h3><a></a>Phase 7: Maintenance:</h3>
<p>Once the system is deployed, and customers start using the developed system, following 3 activities occur</p>
<ul>
<li>Bug fixing - bugs are reported because of some scenarios which are not tested at all</li>
<li>Upgrade - Upgrading the application to the newer versions of the Software</li>
<li>Enhancement - Adding some new features into the existing software</li>
</ul>
<p>The main focus of this SDLC phase is to ensure that needs continue to be met and that the system continues to perform as per the specification mentioned in the first phase.</p>
