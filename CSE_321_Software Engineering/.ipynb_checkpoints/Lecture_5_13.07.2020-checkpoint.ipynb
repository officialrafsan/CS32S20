In these models waterfall, spiral and agile models are more important for us.

Waterfall model is like classical SDLC model or predecessor of SDLC. Waterfall model is linear/sequential. We don't have to learn the definition of the steps rather we have to learn descriptions of the steps. We call it linear cause we first finish the current step we are in then we move on to next step, parallel processing isn't possible here. We use this model to build small size systems like coffeshop management system etc.

1. In **Requirements definition** we collect info about what the user wants. We mainly collect requirement of the user.
2. We design the software and system here like UML or schema diagram.
3. We complete implementation (coding part)  and do unit testing in this section. Unit testing is testing a particular part of the system like testing database, login system etc.
4. In integration and system testing we deliver the system to user and test the whole system in this part. In this section we make sure the software runs perfectly.
5. Operation and maintenance is the longest phase in this life cycle. In this step after delivering the system we checkup the system after certain periods of time. 

![image.png](figures/fig39.PNG)

We use waterfall model when the requirements doesn't change much. We can't use this to make large system. Requirements must be clear to use this model and there should a stable environment. We can't use technologies that changes frequently. The main drawback of this model is we can't take much customer reaction/feedback, there's no scope for risk management. Overlapping process can't happen to move on to the next step you must finish the current one.
