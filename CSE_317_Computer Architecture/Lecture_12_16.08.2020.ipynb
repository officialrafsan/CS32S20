Version 3 is more optimized than version 2. In version 3 we use only 2 variables. Multiplicand and Product, where in MIPS multiplicand is 32 bit and product is 64 bit( 2 * size of Multiplicand). In version 3 we keep the multiplier's value in the right/lower half of the product register and execute the algo same as version 2. In all the version we may ask how much iteration we should do? Iteration number is equal to size of multiplier. In the example below the multiplier is of 4 bit that's why we should do 4 iterations to get the result. And if the iteration number is given explicitly in the exam then we should execute that number of iterations.
