Here calculate the execution time for both compilers using above mentioned method. Now you'll get MISP by dividing instruction by execution time. $\frac{I}{Execution Time}$ So you'll get MIPS for both compiler. Let's assume MIPS for compiler1 is 6 and MIPS for compiler2 is 3. Now you can get a ration by $\frac{MIPS compiler1}{MIPS compiler2}=\frac{6}{3}=2$ So we can see that $\frac{MIPS compiler1}{MIPS compiler2}=2$
that is $MIPS compiler1 = 2* MIPS compiler2$. We can interpret it as according to MIPS compiler1 is 2 times faster than compiler2. Here MIPS means million instructions per second. Higher the MIPS greater the performance. like if MIPS=6 that means the system can execute 6 million instructions per second. That's why increase in MIPS results in increase in performance.
