Execution time (second/program) = cycle per program(cycles/program) x cycle time(seconds/cycle).

in another words,

Cpu execution time for a program = cpu clock cycles for a program * clock cycle time

or 

Cpu execution time for a program = cpu clock cycles for a program * (1/clock rate) = cpu clock cycles for a program / clock rate

Clock rate is measured in Hz its the frequency of cpu. We multiply by 10^9 to convert seconds into nanoseconds. eqn is cycle/second. In the example the clock rate was converted to cycle time. So basically cycle time is inverse of clock rate.

Fetch cycle is same but depending on the type of instruction the execution time changes. Lets consider two instruction , A=B+C and X=y+j[2] . Among these two instructions the second instruction will take more time to execute cause in the first instruction the values are in register and for the 2nd instruction first the value must be fetched from memory then it can be computed. This is how we can say different instructions take different amounts of cycle time.

![image.png](figures/fig21.PNG)

We must have a common architecture for hardware design for common set of instruction. Thus all the instruction has it's fixed cycle time. So if we change the cycle time it means changing the hardware design.

Cpu_execution_time_A  = clock_cycles_A/clock_rate_A<br>
-> 10s = clock_cycles_A/400x10^6Hz<br>
-> clock_cycles_A = 10 * 4x10^8 = 4x10^9 cycle/sec<br>


Cpu_execution_time_B = (1.2 x clock_cycles_A)/clock_rate_B<br>
-> 6s = (1.2 x 4x10^9)/clock_rate_B<br>
-> clock_rate_B = (1.2 x 4x10^9)/6 = 8x10^8 Mhz = 0.8 Ghz
