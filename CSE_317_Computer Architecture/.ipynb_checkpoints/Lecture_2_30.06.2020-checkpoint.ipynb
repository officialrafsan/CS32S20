Performance is a measurement by which we can know about how fast is the hardware we have, how much load can the hardware take. This things depends on the points mentioned in the fig.

**Response Time :** Also called execution time. The total time required for the
computer to complete a task, including disk accesses, memory
accesses, I/O activities, operating system overhead, CPU
execution time, and so on.<br>

**Throughput :** Also called bandwidth. Another measure of performance, it is the number of tasks completed per unit time. Basically it's the load that a computer can take.<br>

**CPU Time :** The actual time the CPU spends computing for a specific task.<br>

**CPU clocking :** The clock speed measures the number of cycles your CPU executes per second, measured in GHz.<br>

**Example :** Do the following changes to a computer system increase
throughput, decrease response time, or both?
1. Replacing the processor in a computer with a faster version
2. Adding additional processors to a system that uses multiple
processors for separate tasks—for example, searching the
web<br>Decreasing response time almost always improves throughput.
Hence, in case 1, both response time and throughput are
improved. In case 2, no one task gets work done faster, so only
throughput increases.
If, however, the demand for processing in the second case was
almost as large as the throughput, the system might force
requests to queue up. In this case, increasing the throughput
could also improve response time, since it would reduce the
waiting time in the queue. Thus, in many real computer systems,
changing either execution time or throughput often affects the
other.



<h2>Let's talk about Execution time or CPU execution time or Response time or Elapsed Time :</h2>

![image.png](figures/fig15.PNG)

From this fig we can see that execution or elapsed time is cpu+wait time. It counts everything. **So what's CPU & wait time?**<br>

**CPU Time :** The actual time the CPU spends computing for a specific task. CPU
time can be further divided into the CPU time spent in the program, called **user
CPU time**, and the CPU time spent in the operating system performing tasks on
behalf of the program, called **system CPU time.** Like i run a program the total time it takes is cpu time, in the cpu time the time it takes to just execute the block of program including taking input from input device is user cpu time and amount of time the processor worked on operating system's functions connected to that specific program is system cpu time for example, after a thread issued a system call.<br>

**Wait Time :** When we run a program if it needs to fetch data from memory (RAM,HDD) then the total time the processor waits to fetch the data is called wait time.

Total execution time : CPU time + Wait time
