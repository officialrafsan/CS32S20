**Arithmetic Instructions**

MIPS instruction have 3 operands. So maximum and minimum operands are 3. Not more or not less than 3. Compiler assigns variables to registers. A = B + C -> add (opcode) then there are 3 registers.


![image.png](figures/fig39.PNG)

![image.png](figures/fig40.PNG)

Operands must be in register to store the values. MIPS has **32 registers**. Each register can hold upto 5bits. that is 0 (00000)->31 (11111).

![image.png](figures/fig41.PNG)

**Memory Reference Instruction**

C = A + B this is register reference cause these variable can be directly stored in registers. but C = A[2] + B here A is an array so this refers to memory. Array is stored in memory cause register may not have sufficient storage to store it. So array is stored in memory and when needed the data of an index can be continuously loaded to register for computation. When we store each value of array from register to memory it's called write operation and when do the opposite it's called read operation. So MIPS is a load & store architecture. MIPS has `two memory reference instruction` one is lw (memory read operation) and another one is sw (memory write operation).

![image.png](figures/fig42.PNG)

![image.png](figures/fig43.PNG)

using lw we load a word. a word is 4bytes = 32bit. Thats why registers correspondingly hold 32bits of data. registers are 32 bit long. Each register can hold 32 bits (4 consecutive 8bit indexes) of data. Data is also 32 bits. That's why instead of going like 0->1->2->.... it goes like 0->4->8->....

![image.png](figures/fig44.PNG)

As mips transfers 4bytes = 32bit at a time during lw or sw operation that's why offset is always 32.
