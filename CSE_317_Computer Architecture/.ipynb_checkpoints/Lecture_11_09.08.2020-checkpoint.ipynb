There are 3 version of a multiplication operation. With each new version hardware is more optimized to execute the same multiplication operation. Let's start with the first version. 

Algo: 

MIPS is 32 bit, in version 1 we check the right most bit of the multiplier(MSB). Depending on the value of MSB(0 or 1) we do certain task. If the MSB is 1 then we add multiplicand and product and store it in the product register. then we execute step 2 and 3. after step 3 MSB changes as right shift happens so we do this again. We do total 32 iteration as MIPS is 32 bits. But in exam we may  have to do 2-4 iteration. This will be mentioned in exam.

In the example though the all the variables are four bits but we choose to take double the bit of multiplier in both multiplicand and product. That's why Multiplier is 4 bits and Multiplicand and product is 8 bits.

![image.png](figures/fig77.PNG)

![image.png](figures/fig78.PNG)

In the example above we only saw simulation for a 4 bit value. But MIPS has 32 bit this results in increased time complexity. Hardware isn't minimized so we need more clock cycle to do the task. But if the hardware was minimized then it would take less clock cycle to do the task. As we take double the bit size of multiplier in multiplicand thus almost all the time upper half of the multiplicand remains 0. So instead of shifting multiplicand to left we can shift product to right. So all of these improvements will be done in version 2.
