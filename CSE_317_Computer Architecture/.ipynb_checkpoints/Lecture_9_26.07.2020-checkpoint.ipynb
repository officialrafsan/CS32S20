` Array is memory reference and variable is register reference.`

In R-type we saw how to do arithmetic calculation using machine language. To calculate this we need to load the values from memory to registers and after operation we need to store the result back to memory. So this load and save operation is done using I-Type instruction. It's a memory reference type instruction. Let's take `lw $t0, 1002($s2)` this example here we are doing a load operation. where destination is `$t0$` and source is `1002($s2)$`where 1002 is the offset and s2 is the base address. If we add the base address with offset then we get the source from where we are to fetch the data and load it. So this is T-type/data transfer instruction. in load operation the **1st operand is destination** but in save operation **1st operand is source**, for example `sw $t0, 1002($s2)` if we had written the instruction like this it means that `$t0$` is source and destination is `1002($s2)$`. 

In this instruction op is 100011 and `rt` is source or Destination (depending on if it's lw{load} or sw{save} operation) `rs` is base address and there is 16bit offset address. Generally offset value is large so it won't fit in 8bits that's why it is 16bit.

I-Type instruction is data transfer instruction. When fetching value from array first it needs to be loaded to a variable then the operation may start.
