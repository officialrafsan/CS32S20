Today's Agenda is to learn the following functions.

`mediocre functions`
- is_displayed() : boolean function to check if an element is displayed
- is_enabled() : boolean function to check if an element is enabled

`important functions`

- is_selected() : boolean function to check if a checkbox/radio button is selected or not
- send_keys() : to write on a text box

```python
#Modules
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium import webdriver
import time
```

```python
# open webpage 
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path='chromedriver.exe', options=options)
driver.set_window_size(1120, 1000)
url = 'https://scholar.google.com/'
driver.get(url)
```

```python
# Find out if the search box is displayed or not in the webpage. If True then search box is displayed.

search_box = driver.find_element_by_xpath('.//*[@id="gs_hdr_tsi"]')
search_box.is_displayed()
```

```python
# Find out if the search box is enabled or not in the webpage. If True then search box is enabled and you can use it.

search_box.is_enabled()
```

```python
# Selecting a radio button. Case law

case = driver.find_element_by_xpath('.//*[@id="gs_hp_sdt2"]')
case.is_selected()
```

Here we can see that Case law isn't selected. That's why when we checked it with is_selected() function we got False.

![image.png](figures/fig1.PNG)

```python
# We know that case law is unselected. Now let's write code to select the case law.

if case.is_selected() == False:
    case.click()
    time.sleep(3)
```

Now case law is selected.

![image.png](figures/fig2.PNG)

```python
# We have selected case laws. Now we need to search about a case. Let's implement it.

search_box.send_keys('murder')
time.sleep(3)
```

murder is now pasted in search box

![image.png](figures/fig3.PNG)

```python
# murder is pasted in search box but we need to click on the search button to search it. Let's do it here 
search = driver.find_element_by_xpath('.//*[@id="gs_hdr_tsb"]/span/span[1]')
search.click()
```

Now we can see the search result
