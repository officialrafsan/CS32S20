```octave
% Newton

ref =117; 

f = @(x) (x^3-7/x+2);
fprime = @(x) (3*x^2+(7/(x+2)^2));

iter = 2;
x_old = 3;
matrix = zeros(10,4);
matrix(1,1) = 0;
matrix(1,2) = 3;
matrix(1,3) = NaN;
matrix(1,4) = NaN;

while iter<=11
    x_new = x_old - f(x_old)/fprime(x_old);
    matrix(iter,1) = iter-1;
    matrix(iter,2) = x_new;
    
    eror = abs((abs(x_new-x_old)/x_new)*100);
    
    matrix(iter,3) = eror;
    sig = 0;
    if eror>5
        sig=0;
    endif

    if eror>0.5 && eror<=5
        sig=1;
    endif

    if eror>0.05 && eror<=0.5
        sig=2;
    endif

    if eror>0.005 && eror<=0.05
        sig=3;
    endif

    if eror>0.0005 && eror<=0.005
        sig=4;
    endif

    if eror>0.00005 && eror<=0.0005
        sig=5;
    endif

    if eror>0.000005 && eror<=0.00005
        sig=6;
    endif
    matrix(iter,4) = sig;
    
    x_old = x_new;
    iter=iter+1;
endwhile

disp(matrix)
    
    
```

```octave
%Bisection

x_lower = 3;
x_upper = 4;
myFunction = @(x) (x/2)-(x+1)^(1/3);

matrix = zeros(10,4);


if myFunction(x_lower)*myFunction(x_upper)<0
    x_mid = (x_lower+x_upper)/2;
    matrix(1,1) = 1;
    matrix(1,2) = x_mid;
    matrix(1,3) = NaN;
    matrix(1,4) = NaN;
    iter = 2;
    
    while iter<11;
        sig=0;
        matrix(iter,1) = iter;
        x_old_mid = x_mid;
         if myFunction(x_mid)*myFunction(x_upper)<0
             x_lower=x_mid;
         else 
             x_upper=x_mid;
         endif
        x_mid = (x_lower+x_upper)/2;
        matrix(iter,2) = x_mid;
        eror = (abs((x_mid - x_old_mid)/x_mid))*100;
        matrix(iter,3) = eror;
        
        
        if eror>0.5 && eror<=5
            sig=1;
        endif

        if eror>0.05 && eror<=0.5
            sig=2;
        endif

        if eror>0.005 && eror<=0.05
            sig=3;
        endif

        if eror>0.0005 && eror<=0.005
            sig=4;
        endif

        if eror>0.00005 && eror<=0.0005
            sig=5;
        endif

        if eror>0.000005 && eror<=0.00005
            sig=6;
        endif
        matrix(iter,4) = sig;
        iter = iter +1;

    endwhile
else
    disp(['There is no root between ', num2str(x_lower), ' and ', num2str(x_upper)])
endif

disp(matrix)
```
