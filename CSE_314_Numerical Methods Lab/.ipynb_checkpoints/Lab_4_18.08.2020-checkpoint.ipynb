```octave
x= -5:1:5;
y=x.^3 + 2*x.^2 - x + 3;

plot(x,y)
title('y')



dydx_num = diff(y) ./diff(x);
dydx_exact = 3*x.^2 + 4.*x -1;
dydx = [[dydx_num,NaN]', dydx_exact']

figure(2)

plot(x,[dydx_num, NaN], x, dydx_exact)

title('dy/dx')

legend('numerical solution', 'analytical solution')
```
