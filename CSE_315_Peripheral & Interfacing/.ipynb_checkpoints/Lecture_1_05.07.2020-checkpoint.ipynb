<div>
<h4>Memory vs. storage</h4>
</div>
<div>
<p>The main difference between computer memory and storage is that memory stores data short-term for immediate access. A computer's moment-to-moment operations are performed using short-term data access &mdash; loading applications, browsing the Web, or editing a spreadsheet. Memory is volatile storage, which means that when you turn the computer off, the data disappears.</p>
<p>Storage (either a hard drive or a solid state drive) stores data long-term for permanent access. It's the component that accesses and stores your files, applications, and operating system. The storage drive is non-volatile, which means the data is stored even if you turn off the computer.</p>
</div>

<p><strong><em>1.What is a serial port?&nbsp;</em></strong></p>
<p><strong>Ans :</strong> In computing, a serial port is an interface that allows a PC to transmit or receive data one bit at a time. The serial port is used for peripherals such as old mice, gaming controllers, modems, and older printers. It is sometimes called a COM port or an RS-232 port, which is its technical name. It is one of the oldest types of interfaces. The serial port is typically the slowest port .</p>
<p>&nbsp;Ex: RS 232<br /><br /></p>
<p><strong><em>2.What is a </em></strong><strong><em>parallel </em></strong><strong><em>port?</em></strong></p>
<p><strong>Ans : </strong>A parallel port is an interface allowing a personal computer (PC) to transmit or receive data down multiple bundled cables to a peripheral device such as a printer.A parallel port is an external interface commonly found on PCs from the early 1980s to early 2000s. It was used to connect peripheral devices such as printers and external storage devices. It was eventually superseded by USB, which provides a smaller connection and significantly faster data transfer rates.</p>
<p>Ex : DB25</p>
<p><strong><em>3.What is a </em></strong><strong><em>USB </em></strong><strong><em>port?</em></strong></p>
<p><strong>Ans </strong><strong>: </strong>USB Stands for "Universal Serial Bus" is a common interface that enables communication between devices and a host controller such as a personal computer (PC). USB is the most common type of computer port used in today's computers. It can be used to connect keyboards, mice, game controllers, printers, scanners, digital cameras, and removable media drives etc.USB is also faster than older ports, such as serial and parallel ports. The USB 1.1 specification supports data transfer rates of up to 12Mb/sec and USB 2.0 has a maximum transfer rate of 480 Mbps.</p>
<p>Ex: Type A<br /><br /></p>
<p><strong><em>4.Relation between peripheral and interface :&nbsp;</em></strong></p>
<p>Peripheral means various components or devices that are connected to cpu. Actually these are I/O devices. </p>
<p>An Interface is a concept that refers to a point of interaction between objects or components and is applicable at the level of both hardware and software.</p>
<p>So the relationship between peripherals and interface basically is interaction. Interface interacts with peripherals to get input from peripheral devices and send them to CPU and upon receiving output from CPU interface again interacts with peripherals to deliver the output.</p>
