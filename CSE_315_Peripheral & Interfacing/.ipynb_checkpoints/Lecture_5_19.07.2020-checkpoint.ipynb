A block of code i.e. finding the maximum in an array, may be used several times in a program. With out writing it multiple time we can separate this block of code from the main program. This separated block is called a function. Usually it is declared outside of the main program and we can use this by just calling the function name and provide parameters if necessary. Here is a formal definition of function. 

![image.png](figures/fig25.PNG)

Prototyping increases readability. Cause you may have 10 functions. Defining them at the beginning isn't good for readability as the setup and loop function falls very below in the program. So by using prototyping we can just declare the name of functions and after the loop function we can define the functions. This increases readability.

If we want to print something to see output we need to use serial port. This port is also used to communicate with arduino. To use serial port we need to open the port. To open it we type Serial.begin(). It takes a parameter that is data rate usually its 9600. This means the port can communicate 9600 bps(bits per sec)
