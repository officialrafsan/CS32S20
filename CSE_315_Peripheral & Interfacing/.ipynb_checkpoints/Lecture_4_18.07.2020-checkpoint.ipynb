We know binary can be in the form of 2,4,8,16. So when we left shift 6 by 2 we get 11000 which is 5 bit. So successor to 5 is 8 that's why we put extra 3 0's in front of 11000 to get 8bit binary 00011000 .

We write code in void loop if we want to continuously run the code when arduino is powered up. And we write code in void setup when we just want to run the code only once.

in digitalWrite we could only supply the pin number and tell that to just turn on and off nothing more than that. But to get a fading effect we use analogWrite this takes pin number and continuous value as parameter. We can see that in the code example the continuous value starts from 0(off) and gradually increases the brightness till 255(maximum brightness). Then we try 255 -> 0 resulting in a fading effect.
