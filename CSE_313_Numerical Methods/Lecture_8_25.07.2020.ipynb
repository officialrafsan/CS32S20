## Direct fit polynomial

![image.png](figures/fig99.PNG)

third order polynomial means the equation will extend to $a_3$ or we can say $t^3$, So the equation will be $v(t)=a_0t^0+a_1t+a_2t^2+a_3t^3$

$->v(t)=a_0+a_1t+a_2t^2+a_3t^3$

From the provided dataset we can generate values of $a_0,a_1,a_2,a_3$. These are polynomial parameters using these we can fit the dataset to polynomial. Now when we are to predict a new value we of `t` we just use the equation.

![image.png](figures/fig100.PNG)

We need to generate values of $a_0,a_1,a_2,a_3$ from given dataset as it is a third order polynomial (we'll learn how to calculate these values, further in this course).Since we want to find the velocity at t=16s, and we are using a third order polynomial, we need to choose the four points closest to t=16s and that also bracket t=16s to evaluate it.The four points are, $t_0=10, t_1=15, t_2=20, t_3=22.5$.

` N.B : Here we are taking two t values before t=16 and after t=16 cause as it's 3rd order polynomial we are considering four t values. But if it was 2nd order polynomial then we need to take three t values. Now the question is should I take two t values before t=16 and one t value after t=16 OR should I take one t value before t=16 and two t values after t=16? Well the answer is you can choose any of these two methods though the error will be different for these two methods but the difference between the error is negligible`
