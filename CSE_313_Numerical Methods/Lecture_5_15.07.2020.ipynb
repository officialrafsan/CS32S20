# Differentiation – Continuous Functions [Read More!](http://mathforcollege.com/nm/mws/gen/02dif/mws_gen_dif_txt_continuous.pdf)

Differentiation is continuous function. In differentiation we'll learn 3 topics enlisted below.

1. Forward Difference Approximation(FDD)
2. Backward Difference Approximation(BDD)
3. Center Difference Approximation(CDD)

![image.png](figures/fig58.PNG)

The above is the forward divided difference approximation of the first derivative. It is called forward because you are taking a point ahead of X . From point X we want to goto a new point that is greater than X and the difference between these points is deltaX. But we don't know the value of new point that's why to know it we write f(X+deltaX).
