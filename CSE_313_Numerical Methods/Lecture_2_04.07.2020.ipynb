# Error measures : 

![image.png](figures/fig8.PNG)

* Evaluate the outcome
* Like when we do iterations if we don't have a stopping criteria it will run forever. Error measuring gives a stopping criteria. Like after reaching a certain iteration the error increases so the iteration stops at that certain level. 

Iteration works like when we have a function f(x) = (x^2+x)/2 ; 0<x<9, like this the range of x is 1 to 8. Lets use 1 for the first iteration then we get a value of f(x) suppose thats 'y', then we put the new value in the equation like f(y) this is the second iteration. The value we put in the equation must be within the range.


![image.png](figures/fig9.PNG)

True error denoted by Et is the difference between the true value (also called the exact
value) and the approximate value. 

**True error = given value {f(x)} - predicted value {f'(x)}**


![image.png](figures/fig10.PNG)

![image.png](figures/fig11.PNG)

![image.png](figures/fig12.PNG)

We take the absolute value of Et that is |Et| = |-0.722|=0.722.

![image.png](figures/fig14.PNG)

The magnitude of true error does not show how bad the error is. A true error of Et = -0.722
may seem to be small, but if the function given in the Example 1
were f(x) = 7x10^-6e^(0.5x) the true error in calculating f'(2) with h=0.3  would be
Et = -0.75061x10^-6. This value of true error is smaller, even when the two problems are
similar in that they use the same value of the function argument, x=2 and the step size,
h=0.3 . This brings us to the definition of relative true error. 


![image.png](figures/fig13.PNG)

Relative true error is defined as the ratio between the true error and the true value.
