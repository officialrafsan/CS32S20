# Today we'll learn source of errors. 

There are two source of errors in numerical method.
1. Round off error
2. Truncation error


Here in the three examples the values are continuous like 0.33333....... but for the sake of calculation we take 0.33 or 0.333 this is called rounding the decimal value to some extent. As we are rounding the value we are loosing significant digits. This is how error occurs due to rounding. The error may be very small but there's still an error. So round off is a source of errors. 

![image.png](figures/fig35.PNG)


To minimize round of error we need to carry all the significant digits in our calculation. Lets see the effects of carrying significant digits in calculations

From this table we can see that when we calculate with most of the significant values the predicted value of alpha is quite same as the given value but when we round it off then we can see certain amount of error. So following this method we can minimize the error.

Truncation error occurs when we change the mathematical procedure of finding approximate value or error generated from mathematical procedures. Like when we do some prediction/approximation each time the mathematical model gives us some error. So we can say this error is generated by the mathematical model.

![image.png](figures/fig40.PNG)

Let's see an example. Here we can see Maclaurin series equation. But this eqn tends to infinity. We can't calculate eX by taking all the values as it's infinite. So we change the model and take 1st 3 terms those are 1+x+(x^2)/2! but this won't be equal to eX. So the difference would be the truncation error. Truncation error = eX - 1+x+(x^2)/2! . This error occurred due to change in mathematical model. 

![image.png](figures/fig41.PNG)

Here in this example we can see that it takes at least 6terms to redure relative approximation error below 1%. Approximate error = present approximation - previous approximation. relative approximation% = (approximate error / present approximation)x100 .  now there's a question how many term required to get at least 1 significant digit. When m=1 then we get |Ea|% < 0.5x10^2-1 = |Ea|% < 5. From the table we can see that for 5,6,... terms |Ea|% < 5. So at least we need 5terms to get at least 1 significant digit.
