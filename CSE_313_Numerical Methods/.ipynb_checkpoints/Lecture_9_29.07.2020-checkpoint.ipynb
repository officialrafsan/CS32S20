## Bisection Method

Bisection method and binary search method are the same thing. Using bisection we need to find the root. So what is this **root?** Root is a point/position where a value/line of y intersects x axis. So we take two values from two corner and try to find the root using binary search. $x_l$ and $x_u$ are low and high points. A function should be continuous, real, changes sign to implement this method and we take two points such that there will exist at least a single root between them, there can be multiple root. This means if the function satisfies `continuous, real, changes sign` these rules there would be at least a single root and also can be multiple root.

![image.png](figures/fig113.PNG)

So we know that the two points we take $f(x_l)$ and $f(x_u)$ these points must have different signs from each other cause the multiplication of these two points must be less than 0 to implement bisection. But in this figure below both the signs of the points are positive. So we can't conclude that there can't be any root, we can see there are two roots in this figures. So we can conclude that there may or may not be any root between the points. But to know that if there exists a root or not we need to use other method as we can't use bisection in this cause it doesn't satisfies to conditions of bisection.
