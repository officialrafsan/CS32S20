## Today's topic is Newton-Raphson Method

In the fig below, where the curve intercepts x-axis that point is called root. We draw each tangent line to reach root. Let's assume the first tangent line intercepts x-axis at $x_i$ point then the 2nd tangent line will intercept at $x_{i+1}$ point, 3rd tangent line will be at $x_{i+2}$ and so on. So with each iteration we are going near to the root. So to calculate we use the equation $x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$. In bisection we needed to calculate using two points but here in this method we only use one point and draw tangent line on it that intercepts x-axis. This reduces time complexity.

![image.png](figures/fig144.PNG)

Derivation of the newton-raphson equation. 

Here we write $f(x_i)-0$  cause in BDD we learned that we subtract rise of two points. But here tangent line has a point where rise is $f(x_i)$ and another point has rise 0 cause it intercepts x-axis thus the rise/ height value is 0.

$f'(x_i) = \frac{f(x_i)-0}{x_i-x_{i+1}}$  

$f'(x_i) = \frac{f(x_i)}{x_i-x_{i+1}}$ 

$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_i)}$ (1st iteration)

$x_{i+2} = x_{i+1} - \frac{f(x_{i+1})}{f'(x_{i+1})}$ (2nd iteration)

and so on...

Stopping criteria would be given in exam. Some criteria are given below in the order of importance/probability of coming in exam.

1. It can be like do 3 or n iteration and give absolute error. 
2. If absolute error is less than tolerance value then stop you reached root otherwise do iteration
3. Try until you get n significant digits in absolute error. Then you'll have to use significant digit measure method for each iteration.

After 18th iteration (table 1) we reach near root but before that we had negative value and according to the algo when the value is less then 0 then we assume we reached root and stop algo. But here we can see some confusion cause after getting negative value we continue and then we get near to root, it was also possible that the value may never be positive again. So should we stop or continue the algo. This confusion is basically the drawback of newton method, divergence at inflection point.
