## Introduction to Scientific Computing

<h1>Why use numerical methods?</h1>
If a problem can be solved without needing heavy computaion like "Mr.A has 3 pens and Mr.B has 4 pens how many pens do they have together> Total they have 7 pens" this type of problem can be solved easily with exact answer. So this problem is analytically solvable. But when a problem isn't analytically solvable then we use numerical methods to solve the problem with precision. Like solving linear/ non-linear problems, fit some points to curve, good approximation and simple solution. 


<h1>How do we solve an engineering problem?</h1>

To solve an problem first we need a precise description of the problem. Then we can analyse it and come up with a mathmetical model and can simulate the model against dataset to evaluate the model

<h1>Let's look at the Trunnion-Hub-Girder problem and how can Numerical Methods can solve it<br><br>Problem : </h1>


![image.png](figures/fig1.PNG)<br>
To make the fulcrum (Figure 1) of a bascule bridge, a long hollow steel shaft called the trunnion is shrink fit into a steel hub. The resulting steel trunnion-hub assembly is then shrink fit into the girder of the bridge.This is done by first immersing the trunnion in a cold medium such as a dryice/alcohol mixture. After the trunnion reaches the steady state temperature of the cold medium, the trunnion outer diameter contracts. The trunnion is taken out of the medium and slid through the hole of the hub<br>
![image.png](figures/fig2.PNG)<br>
While inserting the trunnion in the hub it expands as it heats up and got stuck in the hub but it was pulled out before it got permanently stuck. This is a major problem cause if the trunnion got stuck permanently then we had to make a new hub and trunnion. The engineers did some precalculation **then why did it get stuck?**<br>
The reason is that when the trunnion was chilled it wasn't shrinked enough to slide through the hub. The outer diameter of the of the trunnion is 12.363" and inner diameter of the of the hub is 12.358" the minimum clearance required to fit the trunnion in hub is 0.015". To calculate the contraction of the trunnion, the thermal expansion coefficient is used at room temperature(80F). Room temp is 80F and dryice/alcohol can reach lowest temp of -108F. In that case the reduction ∆D in the outer diameter
of the trunnion is<br>

So the predicted reduction is 0.01504 whis is greater than 0.015. **So why did it still got stuck?**<br>
